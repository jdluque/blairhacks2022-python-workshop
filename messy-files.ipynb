```python
# Make the messy files
filetypes = ['song', 'note', 'book']
for i in range(1000):
    for ftype in filetypes:
        if ftype == 'song':
            extension = '.mp3'
        elif ftype == 'note':
            extension = '.txt'
        else:
            extension = '.pdf'
        with open(f'messy files/{ftype}_{i}{extension}', 'w') as f:
            pass
```

```python
import os
```

```python
dirs_to_make = ['songs', 'notes', 'books']
for dir_name in dirs_to_make:
    os.mkdir(f'messy files/{dir_name}')
```

```python
for filename in os.listdir('messy files/'):
    if os.path.isdir(f'messy files/{filename}'):
        continue
    ftype, name_and_ext = (filename.split('_'))
    os.rename(f'messy files/{filename}', f'messy files/{ftype}s/{filename}')
```
